```python
import os
import numpy as np
from PIL import Image
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import time

# Your paths
TRAIN_FOLDER_PATH = r"D:\\vit study\\Machine Learning\\Gait\\GEI\\GEI\\090-00"
TEST_FOLDER_PATH = r"D:\\vit study\\Machine Learning\\Gait\\GEI\\GEI\\090-01"

print("STEP 5: LOADING ALL DATA WITH COMMON SUBJECTS")
print("="*60)

def get_common_subjects(train_path, test_path):
    """Find subjects present in both train and test folders"""
    
    def extract_subjects(folder_path):
        if not os.path.exists(folder_path):
            return set()
        
        files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
        subjects = set()
        for f in files:
            try:
                subject_id = int(f.split('.')[0])
                subjects.add(subject_id)
            except:
                pass
        return subjects
    
    train_subjects = extract_subjects(train_path)
    test_subjects = extract_subjects(test_path)
    common = train_subjects.intersection(test_subjects)
    
    return sorted(list(common))

def load_all_data_common_subjects(train_path, test_path):
    """Load all data for common subjects only"""
    
    print("Finding common subjects...")
    common_subjects = get_common_subjects(train_path, test_path)
    common_set = set(common_subjects)
    
    print(f"Common subjects found: {len(common_subjects)}")
    print(f"Subject range: {min(common_subjects)} to {max(common_subjects)}")
    
    def load_folder_data(folder_path, target_subjects, folder_name):
        """Load all images for target subjects from a folder"""
        
        if not os.path.exists(folder_path):
            return None, None
        
        print(f"\\nLoading {folder_name} data...")
        
        image_files = [f for f in os.listdir(folder_path) if f.endswith('.png')]
        image_files.sort()
        
        images = []
        labels = []
        
        start_time = time.time()
        processed = 0
        
        for img_file in image_files:
            if processed % 2000 == 0:
                elapsed = time.time() - start_time
                print(f"  Processed {processed}/{len(image_files)} images... ({elapsed:.1f}s)")
            
            try:
                subject_id = int(img_file.split('.')[0])
                
                # Only load if subject is in common subjects
                if subject_id not in target_subjects:
                    processed += 1
                    continue
                
                img_path = os.path.join(folder_path, img_file)
                
                # Load and flatten image
                img = Image.open(img_path)
                img_array = np.array(img, dtype=np.float32)  # Use float32 for efficiency
                img_flat = img_array.flatten()
                
                images.append(img_flat)
                labels.append(subject_id)
                
            except Exception as e:
                print(f"Error processing {img_file}: {e}")
            
            processed += 1
        
        elapsed = time.time() - start_time
        print(f"  Completed {folder_name}: {len(images)} images in {elapsed:.1f}s")
        
        return np.array(images), np.array(labels)
    
    # Load training data
    X_train, y_train = load_folder_data(train_path, common_set, "TRAINING")
    
    # Load test data
    X_test, y_test = load_folder_data(test_path, common_set, "TEST")
    
    return X_train, y_train, X_test, y_test, common_subjects

# Load all data
print("Starting comprehensive data loading...")
print("This may take a few minutes for ~19,000+ images...")

start_total = time.time()

X_train_raw, y_train, X_test_raw, y_test, common_subjects = load_all_data_common_subjects(
    TRAIN_FOLDER_PATH, TEST_FOLDER_PATH
)

total_time = time.time() - start_total

# Display results
print(f"\\n" + "="*60)
print("COMPREHENSIVE DATA LOADING COMPLETE")
print("="*60)

if X_train_raw is not None and X_test_raw is not None:
    print(f"â±ï¸  Total loading time: {total_time:.1f} seconds")
    print(f"\\nðŸ“Š TRAINING DATA:")
    print(f"   Shape: {X_train_raw.shape}")
    print(f"   Memory: {X_train_raw.nbytes / 1024 / 1024:.1f} MB")
    print(f"   Subjects: {len(np.unique(y_train))}")
    print(f"   Subject range: {y_train.min()} to {y_train.max()}")
    
    print(f"\\nðŸ“Š TEST DATA:")
    print(f"   Shape: {X_test_raw.shape}")
    print(f"   Memory: {X_test_raw.nbytes / 1024 / 1024:.1f} MB")
    print(f"   Subjects: {len(np.unique(y_test))}")
    print(f"   Subject range: {y_test.min()} to {y_test.max()}")
    
    print(f"\\nâœ… DATA QUALITY CHECK:")
    print(f"   Train subjects == Test subjects: {set(y_train) == set(y_test)}")
    print(f"   Common subjects count: {len(common_subjects)}")
    print(f"   No missing subjects in dataset: {len(set(y_train)) == len(common_subjects)}")
    
    # Quick normalization
    print(f"\\nðŸ”§ APPLYING NORMALIZATION:")
    print("   Using simple 0-1 scaling (divide by 255)...")
    
    X_train_norm = X_train_raw / 255.0
    X_test_norm = X_test_raw / 255.0
    
    print(f"   Train range: {X_train_norm.min():.3f} to {X_train_norm.max():.3f}")
    print(f"   Test range: {X_test_norm.min():.3f} to {X_test_norm.max():.3f}")
    
    print(f"\\nðŸŽ¯ DATASET READY FOR MODEL TRAINING!")
    print(f"   Features per image: {X_train_raw.shape[1]}")
    print(f"   Total training samples: {len(X_train_raw)}")
    print(f"   Total test samples: {len(X_test_raw)}")
    print(f"   Classification classes: {len(common_subjects)}")

else:
    print("âŒ Error loading data!")

print(f"\\n" + "="*60)
print("STEP 5 COMPLETE - Ready for model training!")
print("="*60)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pickle
import pandas as pd
from sklearn.decomposition import PCA
import seaborn as sns

print("STEP 5.5: DATA VISUALIZATION & DATASET SAVING")
print("="*60)

# Assuming you have the data from Step 5:
# X_train_norm, y_train, X_test_norm, y_test (normalized 0-1)
# X_train_raw, y_train, X_test_raw, y_test (raw 0-255)

print("ðŸ“Š PART 1: VISUALIZING GEI SAMPLES")
print("-" * 40)

def visualize_gei_samples(X_data, y_data, title="GEI Samples", n_samples=12):
    """Visualize sample GEI images"""
    
    # Reshape from flat to 2D for visualization
    img_height, img_width = 128, 88
    
    # Select random samples
    np.random.seed(42)
    indices = np.random.choice(len(X_data), n_samples, replace=False)
    
    fig, axes = plt.subplots(3, 4, figsize=(12, 9))
    fig.suptitle(f'{title} - Random Samples', fontsize=16)
    
    for i, idx in enumerate(indices):
        row, col = i // 4, i % 4
        
        # Reshape flat array back to 2D image
        img = X_data[idx].reshape(img_height, img_width)
        subject_id = y_data[idx]
        
        axes[row, col].imshow(img, cmap='gray')
        axes[row, col].set_title(f'Subject {subject_id}', fontsize=10)
        axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.savefig('gei_samples.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return indices

def compare_train_test_samples(X_train, y_train, X_test, y_test, n_pairs=6):
    """Compare training and test samples for same subjects"""
    
    img_height, img_width = 128, 88
    
    # Find common subjects
    common_subjects = list(set(y_train) & set(y_test))
    selected_subjects = np.random.choice(common_subjects, n_pairs, replace=False)
    
    fig, axes = plt.subplots(2, n_pairs, figsize=(15, 6))
    fig.suptitle('Training vs Test GEI Comparison (Same Subjects)', fontsize=16)
    
    for i, subject_id in enumerate(selected_subjects):
        # Find indices for this subject
        train_idx = np.where(y_train == subject_id)[0][0]
        test_idx = np.where(y_test == subject_id)[0][0]
        
        # Get images
        train_img = X_train[train_idx].reshape(img_height, img_width)
        test_img = X_test[test_idx].reshape(img_height, img_width)
        
        # Plot training image
        axes[0, i].imshow(train_img, cmap='gray')
        axes[0, i].set_title(f'Train\\nSubject {subject_id}', fontsize=9)
        axes[0, i].axis('off')
        
        # Plot test image  
        axes[1, i].imshow(test_img, cmap='gray')
        axes[1, i].set_title(f'Test\\nSubject {subject_id}', fontsize=9)
        axes[1, i].axis('off')
    
    plt.tight_layout()
    plt.savefig('train_test_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

# Visualize random samples from training data
print("\\nVisualizing random training samples...")
train_indices = visualize_gei_samples(X_train_norm, y_train, "Training GEI", n_samples=12)

print("\\nVisualizing train vs test comparison...")
compare_train_test_samples(X_train_norm, y_train, X_test_norm, y_test, n_pairs=6)

print("\\nðŸ“ˆ PART 2: DATA DISTRIBUTION ANALYSIS")
print("-" * 40)

def analyze_pixel_distributions(X_data, title="Pixel Distribution"):
    """Analyze pixel value distributions"""
    
    # Sample some images for analysis
    sample_size = min(1000, len(X_data))
    sample_indices = np.random.choice(len(X_data), sample_size, replace=False)
    sample_data = X_data[sample_indices]
    
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle(f'{title} Analysis', fontsize=16)
    
    # 1. Overall pixel distribution
    all_pixels = sample_data.flatten()
    axes[0, 0].hist(all_pixels, bins=50, alpha=0.7, color='blue', edgecolor='black')
    axes[0, 0].set_title('Overall Pixel Distribution')
    axes[0, 0].set_xlabel('Pixel Value')
    axes[0, 0].set_ylabel('Frequency')
    
    # 2. Mean pixel values per image
    mean_values = np.mean(sample_data, axis=1)
    axes[0, 1].hist(mean_values, bins=30, alpha=0.7, color='green', edgecolor='black')
    axes[0, 1].set_title('Mean Pixel Value per Image')
    axes[0, 1].set_xlabel('Mean Pixel Value')
    axes[0, 1].set_ylabel('Number of Images')
    
    # 3. Standard deviation per image
    std_values = np.std(sample_data, axis=1)
    axes[1, 0].hist(std_values, bins=30, alpha=0.7, color='red', edgecolor='black')
    axes[1, 0].set_title('Std Deviation per Image')
    axes[1, 0].set_xlabel('Standard Deviation')
    axes[1, 0].set_ylabel('Number of Images')
    
    # 4. Average GEI (mean across all samples)
    mean_gei = np.mean(sample_data, axis=0).reshape(128, 88)
    im = axes[1, 1].imshow(mean_gei, cmap='gray')
    axes[1, 1].set_title('Average GEI Pattern')
    axes[1, 1].axis('off')
    plt.colorbar(im, ax=axes[1, 1])
    
    plt.tight_layout()
    plt.savefig('pixel_distribution_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return {
        'mean_pixel_overall': np.mean(all_pixels),
        'std_pixel_overall': np.std(all_pixels),
        'mean_image_means': np.mean(mean_values),
        'mean_image_stds': np.mean(std_values)
    }

print("Analyzing training data distributions...")
train_stats = analyze_pixel_distributions(X_train_norm, "Training Data")

print(f"Training data statistics:")
print(f"  Overall mean pixel: {train_stats['mean_pixel_overall']:.4f}")
print(f"  Overall std pixel: {train_stats['std_pixel_overall']:.4f}")
print(f"  Mean of image means: {train_stats['mean_image_means']:.4f}")
print(f"  Mean of image stds: {train_stats['mean_image_stds']:.4f}")

print("\\nðŸ” PART 3: DIMENSIONALITY REDUCTION VISUALIZATION")
print("-" * 40)

def visualize_pca_projection(X_data, y_data, title="PCA Projection", n_samples=2000):
    """Visualize data in 2D using PCA"""
    
    print(f"Computing PCA for {title}...")
    
    # Sample data for visualization (PCA on full 9K+ samples takes too long)
    if len(X_data) > n_samples:
        indices = np.random.choice(len(X_data), n_samples, replace=False)
        X_sample = X_data[indices]
        y_sample = y_data[indices] 
    else:
        X_sample = X_data
        y_sample = y_data
    
    # Compute PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_sample)
    
    # Create scatter plot
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_sample, 
                         cmap='viridis', alpha=0.6, s=20)
    plt.colorbar(scatter, label='Subject ID')
    plt.title(f'{title} - PCA Projection\\n'
              f'Explained Variance: PC1={pca.explained_variance_ratio_[0]:.3f}, '
              f'PC2={pca.explained_variance_ratio_[1]:.3f}')
    plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
    plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
    
    plt.savefig('pca_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"  PC1 explains {pca.explained_variance_ratio_[0]:.3f} of variance")
    print(f"  PC2 explains {pca.explained_variance_ratio_[1]:.3f} of variance")
    print(f"  Total variance explained: {sum(pca.explained_variance_ratio_):.3f}")
    
    return pca

print("Creating PCA visualization (using 2000 sample points)...")
pca_model = visualize_pca_projection(X_train_norm, y_train, "Training Data GEI")

print("\\nðŸ’¾ PART 4: SAVING DATASET")
print("-" * 40)

def save_dataset_multiple_formats(X_train, y_train, X_test, y_test, base_path="gei_dataset"):
    """Save dataset in multiple formats for easy reuse"""
    
    print("Saving dataset in multiple formats...")
    
    # 1. NumPy format (.npy) - Most efficient for numerical data
    print("  Saving as NumPy files (.npy)...")
    np.save(f'{base_path}_X_train.npy', X_train)
    np.save(f'{base_path}_y_train.npy', y_train) 
    np.save(f'{base_path}_X_test.npy', X_test)
    np.save(f'{base_path}_y_test.npy', y_test)
    
    # 2. Pickle format (.pkl) - Preserves exact Python objects
    print("  Saving as Pickle file (.pkl)...")
    dataset_dict = {
        'X_train': X_train,
        'y_train': y_train,
        'X_test': X_test, 
        'y_test': y_test,
        'shape': X_train.shape,
        'n_subjects': len(np.unique(y_train)),
        'feature_dim': X_train.shape[1],
        'image_shape': (128, 88),
        'normalization': '0-1 scaling',
        'description': 'OUMVLP GEI Dataset - 90 degree view, common subjects only'
    }
    
    with open(f'{base_path}_complete.pkl', 'wb') as f:
        pickle.dump(dataset_dict, f)
    
    # 3. CSV format (for first 100 samples - demonstration)
    print("  Saving sample as CSV (first 100 samples)...")
    sample_size = min(100, len(X_train))
    
    # Create DataFrame with features + labels
    feature_cols = [f'pixel_{i}' for i in range(X_train.shape[1])]
    
    train_df = pd.DataFrame(X_train[:sample_size], columns=feature_cols)
    train_df['subject_id'] = y_train[:sample_size]
    train_df['split'] = 'train'
    
    test_df = pd.DataFrame(X_test[:sample_size], columns=feature_cols)
    test_df['subject_id'] = y_test[:sample_size]  
    test_df['split'] = 'test'
    
    combined_df = pd.concat([train_df, test_df], ignore_index=True)
    combined_df.to_csv(f'{base_path}_sample_100.csv', index=False)
    
    # 4. Create metadata file
    print("  Creating metadata file...")
    metadata = {
        'dataset_name': 'OUMVLP GEI - 90 Degree View',
        'total_subjects': len(np.unique(y_train)),
        'train_samples': len(X_train),
        'test_samples': len(X_test),
        'feature_dimension': X_train.shape[1],
        'image_dimensions': '128x88 pixels',
        'data_type': str(X_train.dtype),
        'normalization': '0-1 scaling (divided by 255)',
        'missing_subjects': '141 subjects missing from original 10,306',
        'train_test_overlap': 'Same subjects in train and test (closed-set evaluation)',
        'file_formats': {
            'numpy': f'{base_path}_X_train.npy, {base_path}_y_train.npy, etc.',
            'pickle': f'{base_path}_complete.pkl', 
            'csv_sample': f'{base_path}_sample_100.csv'
        },
        'loading_instructions': {
            'numpy': 'X_train = np.load("filename.npy")',
            'pickle': 'with open("filename.pkl", "rb") as f: data = pickle.load(f)'
        }
    }
    
    with open(f'{base_path}_metadata.txt', 'w') as f:
        for key, value in metadata.items():
            f.write(f"{key}: {value}\\n")
    
    # Calculate file sizes
    import os
    files_info = {}
    for ext in ['.npy', '.pkl', '.csv', '.txt']:
        for file in os.listdir('.'):
            if file.startswith(base_path) and file.endswith(ext):
                size_mb = os.path.getsize(file) / (1024 * 1024)
                files_info[file] = f'{size_mb:.1f} MB'
    
    print(f"\\n  Files created:")
    for filename, size in files_info.items():
        print(f"    {filename}: {size}")
    
    return files_info

# Save the dataset
print("Saving complete dataset...")
file_info = save_dataset_multiple_formats(X_train_norm, y_train, X_test_norm, y_test)

print(f"\\nâœ… DATASET SAVED SUCCESSFULLY!")
print(f"   Use .npy files for fast loading in future sessions")
print(f"   Use .pkl file for complete dataset with metadata")
print(f"   Check metadata.txt for detailed information")

print(f"\\n" + "="*60)
print("STEP 5.5 COMPLETE - Data visualized and saved!")
print("="*60)
print("\\nREADY FOR NEXT STEPS:")
print("âœ… Data visualization completed")
print("âœ… Dataset saved in multiple formats") 
print("âœ… Ready for model training (Step 6)")
print("âœ… Ready for advanced analysis")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import time
import os

print("STEP 6: CNN FOR GAIT BIOMETRIC RECOGNITION")
print("="*60)

# Load the saved dataset (much faster than reprocessing!)
print("ðŸ“‚ LOADING SAVED DATASET")
print("-" * 30)

# Option 1: Load from NumPy files (fastest)
print("Loading dataset from NumPy files...")
X_train = np.load('gei_dataset_X_train.npy')
y_train = np.load('gei_dataset_y_train.npy') 
X_test = np.load('gei_dataset_X_test.npy')
y_test = np.load('gei_dataset_y_test.npy')

print(f"âœ… Dataset loaded successfully!")
print(f"   Training: {X_train.shape}")
print(f"   Testing: {X_test.shape}")
print(f"   Classes: {len(np.unique(y_train))}")

# Reshape data for CNN (samples, height, width, channels)
print("\\nðŸ”„ PREPARING DATA FOR CNN")
print("-" * 30)

# Reshape from flat (9708, 11264) to image format (9708, 128, 88, 1)
img_height, img_width = 128, 88
X_train_cnn = X_train.reshape(-1, img_height, img_width, 1)
X_test_cnn = X_test.reshape(-1, img_height, img_width, 1)

print(f"Reshaped for CNN:")
print(f"   X_train shape: {X_train_cnn.shape}")
print(f"   X_test shape: {X_test_cnn.shape}")
print(f"   Data type: {X_train_cnn.dtype}")
print(f"   Value range: {X_train_cnn.min():.3f} to {X_train_cnn.max():.3f}")

# Prepare labels for multi-class classification
print("\\nPreparing labels...")

# Create label mapping (subject IDs to continuous indices)
unique_subjects = np.unique(y_train)
subject_to_idx = {subject: idx for idx, subject in enumerate(unique_subjects)}
idx_to_subject = {idx: subject for subject, idx in subject_to_idx.items()}

# Convert subject IDs to indices
y_train_idx = np.array([subject_to_idx[subject] for subject in y_train])
y_test_idx = np.array([subject_to_idx[subject] for subject in y_test])

# Convert to categorical for CNN
y_train_cat = to_categorical(y_train_idx, num_classes=len(unique_subjects))
y_test_cat = to_categorical(y_test_idx, num_classes=len(unique_subjects))

print(f"Label preparation:")
print(f"   Original labels: {y_train.min()} to {y_train.max()}")
print(f"   Mapped to indices: 0 to {len(unique_subjects)-1}")
print(f"   Categorical shape: {y_train_cat.shape}")

print("\\nðŸ—ï¸ BUILDING SIMPLE CNN MODEL")
print("-" * 30)

def create_simple_gei_cnn(input_shape, num_classes):
    """Create a simple CNN for GEI classification"""
    
    model = keras.Sequential([
        # Input layer
        layers.Input(shape=input_shape),
        
        # First convolutional block
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Second convolutional block  
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Third convolutional block
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Fourth convolutional block
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        
        # Flatten and dense layers
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        
        # Output layer
        layers.Dense(num_classes, activation='softmax')
    ])
    
    return model

# Create the model
input_shape = (img_height, img_width, 1)
num_classes = len(unique_subjects)

model = create_simple_gei_cnn(input_shape, num_classes)

# Display model architecture
print("Model Architecture:")
model.summary()

# Compile model
print("\\nâš™ï¸ COMPILING MODEL")
print("-" * 30)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy', 'top_k_categorical_accuracy']  # top-5 accuracy
)

print("âœ… Model compiled with:")
print("   Optimizer: Adam")
print("   Loss: Categorical Crossentropy")
print("   Metrics: Accuracy, Top-5 Accuracy")

print("\\nðŸƒ TRAINING MODEL")
print("-" * 30)

# Training parameters
batch_size = 32
epochs = 20  # Start with fewer epochs
validation_split = 0.1  # Use 10% of training data for validation

print(f"Training parameters:")
print(f"   Batch size: {batch_size}")
print(f"   Epochs: {epochs}")
print(f"   Validation split: {validation_split}")
print(f"   Training samples: {len(X_train_cnn)}")

# Add callbacks for better training
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        min_lr=1e-7
    )
]

print("\\nâ±ï¸ Starting training...")
start_time = time.time()

# Train the model
history = model.fit(
    X_train_cnn, y_train_cat,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=validation_split,
    callbacks=callbacks,
    verbose=1
)

training_time = time.time() - start_time
print(f"\\nâœ… Training completed in {training_time/60:.1f} minutes")

print("\\nðŸ“Š EVALUATING MODEL")
print("-" * 30)

# Evaluate on test set
print("Evaluating on test set...")
start_time = time.time()

test_loss, test_acc, test_top5 = model.evaluate(
    X_test_cnn, y_test_cat,
    batch_size=batch_size,
    verbose=1
)

eval_time = time.time() - start_time

print(f"\\nðŸŽ¯ TEST RESULTS:")
print(f"   Test Loss: {test_loss:.4f}")
print(f"   Test Accuracy (Rank-1): {test_acc:.4f} ({test_acc*100:.2f}%)")
print(f"   Test Top-5 Accuracy: {test_top5:.4f} ({test_top5*100:.2f}%)")
print(f"   Evaluation time: {eval_time:.1f} seconds")

# Make predictions for detailed analysis
print("\\nðŸ” DETAILED PREDICTIONS ANALYSIS")
print("-" * 30)

y_pred_proba = model.predict(X_test_cnn, batch_size=batch_size, verbose=1)
y_pred_idx = np.argmax(y_pred_proba, axis=1)

# Convert back to original subject IDs
y_pred_subjects = np.array([idx_to_subject[idx] for idx in y_pred_idx])
y_true_subjects = y_test

# Calculate accuracy with original subject IDs
accuracy = accuracy_score(y_true_subjects, y_pred_subjects)
print(f"\\nâœ… Final Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

# Top-k accuracy calculation
def calculate_topk_accuracy(y_true_idx, y_pred_proba, k=5):
    """Calculate top-k accuracy"""
    top_k_pred = np.argsort(y_pred_proba, axis=1)[:, -k:]
    correct = 0
    for i, true_idx in enumerate(y_true_idx):
        if true_idx in top_k_pred[i]:
            correct += 1
    return correct / len(y_true_idx)

top5_manual = calculate_topk_accuracy(y_test_idx, y_pred_proba, k=5)
top10_manual = calculate_topk_accuracy(y_test_idx, y_pred_proba, k=10)

print(f"Manual Top-5 Accuracy: {top5_manual:.4f} ({top5_manual*100:.2f}%)")
print(f"Manual Top-10 Accuracy: {top10_manual:.4f} ({top10_manual*100:.2f}%)")

# Training history visualization
print("\\nðŸ“ˆ TRAINING HISTORY")
print("-" * 30)

def plot_training_history(history):
    """Plot training history"""
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # Plot accuracy
    ax1.plot(history.history['accuracy'], label='Training Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True)
    
    # Plot loss
    ax2.plot(history.history['loss'], label='Training Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.savefig('cnn_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

plot_training_history(history)

print(f"\\nðŸ’¾ SAVING MODEL")
print("-" * 30)

# Save model
model.save('gei_cnn_simple.keras')
print("âœ… Model saved as 'gei_cnn_simple.keras'")

print(f"\\n" + "="*60)
print("STEP 6 COMPLETE - Simple CNN trained!")
print("="*60)

print("\\nðŸ“‹ RESULTS SUMMARY:")
print(f"   Architecture: Simple CNN (4 conv blocks)")
print(f"   Parameters: ~{model.count_params():,}")
print(f"   Training time: {training_time/60:.1f} minutes")
print(f"   Final accuracy: {accuracy*100:.2f}%")
print(f"   Top-5 accuracy: {top5_manual*100:.2f}%")

print("\\nðŸ”„ NEXT IMPROVEMENT IDEAS:")
print("   âœ… Baseline CNN established")
print("   ðŸŽ¯ Try deeper architectures (ResNet, DenseNet)")
print("   ðŸ”§ Data augmentation")  
print("   ðŸ“Š Better loss functions (Triplet, ArcFace)")
print("   âš¡ Transfer learning")
```
